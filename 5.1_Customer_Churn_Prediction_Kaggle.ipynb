### Fit and Test Models

```python
## KNN

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(Xtrain.values, ytrain.values)

knn_pred = knn.predict(Xtest.values)

from sklearn.metrics import accuracy_score

knn_accu = accuracy_score(ytest, knn_pred)
knn_accu
```

```python
## Logistic Regression

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(fit_intercept=True)

lr.fit(Xtrain.values, ytrain.values)

# print("Coefficient: " + lr.coef_)
# print("Intercept: " + lr.intercept_)

lr_pred= lr.predict(Xtest.values)

lr_accu = accuracy_score(ytest, lr_pred)
lr_accu
```

```python
## SVM

from sklearn.svm import SVC

svm = SVC(kernel='rbf', C=1).fit(Xtrain.values, ytrain.values)

svm_pred = svm.predict(Xtest.values)

svm_accu = accuracy_score(ytest, svm_pred)
svm_accu
```

```python
## Gaussian Naive Bayes

from sklearn.naive_bayes import GaussianNB

nb = GaussianNB().fit(Xtrain.values, ytrain.values)

nb_pred = nb.predict(Xtest)

nb_accu = accuracy_score(ytest, nb_pred)
nb_accu
```
