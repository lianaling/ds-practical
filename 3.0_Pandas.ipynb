### indexing and selecting data

```python
dates = pd.date_range('1/5/2018', periods=8)
df3 = pd.DataFrame(np.random.randn(8, 4), index=dates, columns=['A', 'B', 'C', 'D'])
df3
```

```python
s = df3['A']
s
```

```python
df3.to_clipboard(excel=True,sep=",")
```

```python
s[dates[5]]
```

`iloc` is integer position-based, either rows or columns

```python
df3.iloc[5]
```

```python
df3[:4]
```

`loc` is label-based

```python
print(df3[2:4])
print(df3.iloc[2:4])

print(df3.loc[:,['B', 'A']])
```

Putting integer position in label-based will cause error

```python
print(df3.loc[2:4])
```

```python
df4 = pd.DataFrame(np.random.randn(6,4),
                   index=list('abcdef'),
                   columns=list('ABCD'))

df4
```

```python
df4.loc[['a', 'b', 'd'], :]
```

```python
df4.loc['d':, 'A':'C']
```

### Note:

loc: selection by label

iloc: selection by position

### Grouping

```python
df5 = pd.DataFrame({'A' : ['foo', 'bar', 'foo', 'bar',
                           'foo', 'bar', 'foo', 'foo'],
                    'B' : ['one', 'one', 'two', 'three',
                           'two', 'two', 'one', 'three'],
                    'C' : np.random.randn(8),
                    'D' : np.random.randn(8)})

df5
```

```python
df5.groupby('A').mean()
```

```python
df5.groupby(['A','B']).mean()
```
